```python
from ultralytics import YOLO
import cv2
```

```python
yolo_pose = YOLO("yolov8n-pose.pt")
```

```python
vid = cv2.VideoCapture(0)

while True:
    _, frame = vid.read()
    result = yolo_pose(frame, classes=(0, ))[0]
    if result:
        frame = result.plot()
        for keypoints in result.keypoints.xy:
            message = "Face is Visible"
            for i in range(5):
                x, y = keypoints[i].int().tolist()
                if x==0 and y==0:
                    message = "Face is Not Visible"
                    break
        cv2.putText(frame, message, keypoints[2].int().tolist(), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 1)
    cv2.imshow('output', frame)
    
    if cv2.waitKey(1) & 0xFF in (ord('q'), ord('Q')):
        break
    
vid.release()
cv2.destroyAllWindows()
```
